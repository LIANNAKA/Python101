```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
df.info()
```

```python
plt.xlabel('region')
plt.ylabel('charges')
plt.scatter(df.region, df.charges, color='red', marker='+')
```

```python
plt.xlabel('sex')
plt.ylabel('charges')
plt.scatter(df.sex, df.charges, color='red', marker='+')
```

```python

plt.figure(figsize=(6,4))
sns.scatterplot(x='age',y='charges',data=df)
plt.title('Age vs charges')
plt.show()
```

```python
#  Correlation between age and charges
correlation = df['age'].corr(df['charges'])
print('Correlation between age and charges:',correlation)
```

```python
#  correlation for non-smokers with charges
non_smokers = df[df['smoker']=='no']
print("Non-smoker correlation:", non_smokers['age'].corr(non_smokers['charges']))
```

```python
#  correlation for smokers with charges
smokers = df[df['smoker']=='yes']
print("smoker correlation:", smokers['age'].corr(smokers['charges']))
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="bmi", y="charges", data=df)
plt.title("BMI vs Charges")
plt.show()

print("Correlation BMI & Charges:", df["bmi"].corr(df["charges"]))
```

```python
# BMI correlation for non-smokers
non_smokers = df[df["smoker"] == "no"]
print("Non-smoker BMI correlation:", non_smokers["bmi"].corr(non_smokers["charges"]))

# BMI correlation for smokers
smokers = df[df["smoker"] == "yes"]
print("Smoker BMI correlation:", smokers["bmi"].corr(smokers["charges"]))
```
