```python
import mysql.connector
```

```python
# create connection to the database

mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Lia@mysql123"
)

print(mydb)
```

```python
#  create database
mycursor = mydb.cursor()
mycursor.execute("CREATE DATABASE mydatabase")
```

```python
# create a table, make sure to mention the database name when creating
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Lia@mysql123",
    database="mydatabase"
)

mycursor = mydb.cursor()
mycursor.execute("CREATE TABLE customers (name VARCHAR(255), address VARCHAR(255))")
```

```python
#  check if the table exists
mycursor.execute("SHOW TABLES")
for x in mycursor:
    print(x)
```

```python
# primary key
mycursor.execute("CREATE TABLE customers (id INT AUTO_INCREMENT PRIMARY KEY, name VARCHAR(255), address VARCHAR(255))")
```

```python
mycursor.execute("ALTER TABLE customers ADD COLUMN id INT AUTO_INCREMENT PRIMARY KEY")
```

```python
#  INSERT INTO
sql = "INSERT INTO customers (name, address) VALUES (%s, %s)"
val = ("John", "highway 21")
mycursor.execute(sql, val)

mydb.commit() # it is required to make the changes, otherwise no changes are made to the table

print(mycursor.rowcount, "record inserted.")
```

```python
#  insert multiple rows
sql = "INSERT INTO customers (name, address) VALUES (%s, %s)"
val = [
  ('Peter', 'Lowstreet 4'),
  ('Amy', 'Apple st 652'),
  ('Hannah', 'Mountain 21'),
  ('Michael', 'Valley 345'),
  ('Sandy', 'Ocean blvd 2'),
  ('Betty', 'Green Grass 1'),
  ('Richard', 'Sky st 331'),
  ('Susan', 'One way 98'),
  ('Vicky', 'Yellow Garden 2'),
  ('Ben', 'Park Lane 38'),
  ('William', 'Central st 954'),
  ('Chuck', 'Main Road 989'),
  ('Viola', 'Sideway 1633')
]

mycursor.executemany(sql, val)

mydb.commit()

print(mycursor.rowcount, "was inserted.")
```

```python
#  get inserted id
sql = "INSERT INTO customers (name, address) VALUES (%s, %s)"
val = ("Michelle", "Blue Village")
mycursor.execute(sql, val)

mydb.commit()

print("1 record inserted, ID:", mycursor.lastrowid)
```

```python
# select from table
mycursor.execute("SELECT * FROM customers")
myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

```python
#  select columns
mycursor = mydb.cursor(buffered=True)

mycursor.execute("SELECT name, address FROM customers")
myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

```python
#  select only one row

mycursor.execute("SELECT * FROM customers")

myresult = mycursor.fetchone()

print(myresult)
```

```python
mycursor = mydb.cursor(buffered=True)

mycursor.execute("SELECT * FROM customers")
mycursor.fetchall()   # clear previous result

sql = "SELECT * FROM customers WHERE address='Mountain 21'"
mycursor.execute(sql)

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

```python
#  filtering with WHERE
sql = "SELECT * FROM customers WHERE address ='Mountain 21'"
mycursor = mydb.cursor(buffered=True)

mycursor.execute(sql)

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

```python
sql = "SELECT * FROM customers WHERE address LIKE '%way%'"
mycursor = mydb.cursor(buffered=True)

mycursor.execute(sql)

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

```python
#  prevent sql injection
mycursor = mydb.cursor(buffered=True)
sql = "SELECT * FROM customers WHERE address = %s"
adr = ("Yellow Garden 2",)

mycursor.execute(sql, adr)

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

```python
#  order by
mycursor = mydb.cursor(buffered=True)
sql = "SELECT * FROM customers ORDER BY name"

mycursor.execute(sql)

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

```python
#  descending order
sql = "SELECT * FROM customers ORDER BY name DESC"

mycursor.execute(sql)

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

```python
#  delete from by
sql = "DELETE FROM customers WHERE address = 'Mountain 21'"

mycursor.execute(sql)

mydb.commit()

print(mycursor.rowcount, "record(s) deleted")
```

```python
#  prevent sql injection
mycursor = mydb.cursor(buffered=True)
sql = "DELETE FROM customers WHERE address = %s"
adr = ("Yellow Garden 2", )

mycursor.execute(sql, adr)

mydb.commit()

print(mycursor.rowcount, "record(s) deleted")
```

```python
#  drop table
sql = "DROP TABLE customers"

mycursor.execute(sql)
```

```python
#  drop only if exist
sql = "DROP TABLE IF EXISTS customers"

mycursor.execute(sql)
```

```python
# update table
mycursor = mydb.cursor(buffered=True)
sql = "UPDATE customers SET address = 'Canyon 123' WHERE address = 'Valley 345'"

mycursor.execute(sql)

mydb.commit()

print(mycursor.rowcount, "record(s) affected")
```

```python
#  prevent sql injection
mycursor = mydb.cursor(buffered=True)
sql = "UPDATE customers SET address = %s WHERE address = %s"
val = ("Valley 345", "Canyon 123")

mycursor.execute(sql, val)

mydb.commit()

print(mycursor.rowcount, "record(s) affected")
```

```python
#  limit
mycursor.execute("SELECT * FROM customers LIMIT 5")

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

```python
#  start from other position
mycursor = mydb.cursor(buffered=True)
mycursor.execute("SELECT * FROM customers LIMIT 5 OFFSET 2")

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```
