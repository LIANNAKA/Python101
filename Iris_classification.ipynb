```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris_data = load_iris()
iris_data
```

```python
iris_data = load_iris()
iris_df = pd.DataFrame(data=iris_data['data'], columns=iris_data['feature_names'])
iris_df['species'] = iris_data['target_names'][iris_data['target']]
iris_df
```

```python
import matplotlib.pyplot as plt
plt.scatter(iris_df.species, iris_df['sepal length (cm)'],color='red', marker='+')
```

```python
X = iris_df.drop('species', axis=1)
y = iris_df.species
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
dummies = pd.get_dummies(iris_df['species'], dtype=int)
dummies
```

```python
df_dummies = pd.concat([iris_df,dummies], axis="columns")
df_dummies
```

```python
df_dummies.drop(['species'],axis="columns", inplace=True)
df_dummies
```

```python
model.fit(X_train,y_train)
```

```python
len(X_train)
```

```python
model.predict(X_test)
```

```python
len(y_train)
```

```python
model.score(X,y)
```
