```python
import numpy as np

# Dataset
X = np.array([1,2,3,4])
y = np.array([2,4,6,8])

# Initialize parameters
w = 0.0
b = 0.0

learning_rate = 0.01
epochs = 1000

n = len(X)

for i in range(epochs):

    # Prediction
    y_pred = w * X + b

    # Gradients
    dw = -2 * np.mean(X * (y - y_pred))
    db = -2 * np.mean(y - y_pred)

    # Update
    w = w - learning_rate * dw
    b = b - learning_rate * db

print("Final w:", w)
print("Final b:", b)
```
