```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_technology_companies_by_revenue'

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/120.0 Safari/537.36"
}

page = requests.get(url, headers=headers)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = 'wikitable sortable plainrowheads static-row-numbers')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
Companies = table.find_all('th')
```

```python
Companies
```

```python
Comapany_table_titles = [title.text.strip() for title in Companies]

print(Comapany_table_titles)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(columns = Comapany_table_titles)
df
```

```python
df.columns = df.columns.str.replace(r'\[1\]', '', regex=True)
```

```python
for col in ['Revenue ($B) USD', 'Profit ($B) USD']:
    df[col] = pd.to_numeric(
        df[col]
        .astype(str)
        .str.strip()                      # remove spaces
        .str.replace(r'[^\d.-]', '', regex=True),  # keep only numbers, . and -
        errors='coerce'
    )


```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df['Revenue ($B) USD'] = df['Revenue ($B) USD'].str.replace('$', '', regex=False).astype(float)

df['Profit ($B) USD'] = df['Profit ($B) USD'].str.replace('$', '', regex=False).astype(float)
```

```python
df.columns = df.columns.str.strip()
```

```python
# Sorting the tech companies by Revenue($B) USD
df = df.sort_values(by='Revenue ($B) USD', ascending=False)
```
