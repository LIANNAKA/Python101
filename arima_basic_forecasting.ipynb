```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Sample time series data
data = [10, 12, 13, 15, 18, 20, 23, 25]

# Convert to series
series = pd.Series(data)

# Build ARIMA model
model = ARIMA(series, order=(1,1,1))

# Train model
model_fit = model.fit()

# Forecast next 3 values
forecast = model_fit.forecast(steps=3)

print(forecast)
```

```python
import matplotlib.pyplot as plt
# Plot original data
plt.plot(series, label='Original Data')

# Plot forecast
plt.plot(range(len(series), len(series)+3),
         forecast,
         label='Forecast')

plt.legend()
plt.show()
```
